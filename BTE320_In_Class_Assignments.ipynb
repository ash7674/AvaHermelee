Append items to list iteratively (while loop). starting with an empty list, design a program in python that populates the list by repeatedly asking the user for numerical inputs. the process stops when the user gives 0. finally, the list should be displayed in descending order.

```python
numbers = []  # start with an empty list

while True:
    num = float(input("Enter a number (0 to stop): "))

    if num == 0:
        break
    numbers.append(num)

numbers.sort(reverse=True)

print("Numbers in descending order:", numbers)
```

dynamic dictionary; N distinct key:value pairs (while loop/ if-elif-else statements). propulate an empty dictionary iteratively N times (choose your own value for N) with name:count pairs. Names (keys) are strings and counts (values) are integers. the keys must be given as inputs from the user on each iteration. use conditionals to check if the key has been added in the dictionary before. if not, set its count to 1; otherwise, increase its count by 1.

```python
names = {}

while True:
  name = input("Give a name: ")
  if name == "":
    break

  if name in names:
    names[name] += 1
  else:
    names[name] = 1

print(names)
```

```python
def roi(cost, revenue):

    assert cost != 0
    assert isinstance(cost, (int, float))
    assert isinstance(revenue, (int, float))

    return (revenue - cost) / cost

cost = float(input("Enter cost: "))
revenue = float(input("Enter revenue: "))

print(roi(cost, revenue))
```

```python
def loan_payment(principal, annual_rate, months):

    assert isinstance(principal, (int, float))
    assert isinstance(annual_rate, (int, float))
    assert isinstance(months, int)

    if principal <= 0:
        raise ValueError("Invalid input values")
    try:
        r = annual_rate / 12
        if r == 0:
            pmt = principal / months
        else:
            pmt = r * principal / (1 - (1+r) ** (-months))
    except Exception as e:
        raise Exception(f"Error: {e}")

    return pmt
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx = (self.x - other.x)**2
        dy = (self.y - other.y)**2
        dist = (dx + dy) ** 0.5
        return dist


class Rectangle:
    def __init__(self, x1, y1, x2, y2, x3, y3, x4, y4):
        self.p1 = Point(x1,y1)  # upper left
        self.p2 = Point(x2,y2)  # upper right
        self.p3 = Point(x3,y3)  # lower left
        self.p4 = Point(x4,y4)  # lower right
        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height
```

```python
square = Rectangle(0,3,3,3,0,0,3,0)
square.area()
```

```python
class Employee:
    def __init__(self, name, hourly_wage, hours):
        self.name = name
        self.__hourly_wage = hourly_wage
        self.hours = hours

    def get_wage(self):
        return self.__hourly_wage

    def set_wage(self, new_wage):
        assert newWage > 0
        self.__hourly_wage = newWage

    def calculate_pay(self):
        return self.__hourly_wage * self.hours

    def __str__(self):
        return f"Employee Name: {self.name}\nHourly Charge: {self.__hourly_wage}$"


class Manager(Employee):
    def __init__(self, name, hourly_wage, hours, bonus):
        super().__init__(name, hourly_wage, hours)
        self.bonus = bonus  # float between 0 - 1; e.g., 0.2 means 20% extra

    def calculate_pay(self):
        return (1 + self.bonus) * super().get_wage() * self.hours


employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)

print(employee)
print(manager)
print()

print(f"Employee {employee.name} weekly salary: {employee.calculate_pay()}$.")
print(f"Manager {manager.name} weekly salary: {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```

```python
class Student:
    def __init__(self, name, number):
        self.__name = name
        self.number = number
        self.__courses = []

    def get_courses(self):
        return self.__courses

    def get_name(self):
        return self.__name

    def set_name(self, newName):
        self.__name = newName

    def __len__(self):
        return len(self.__courses)

    def enroll(self, course):
        if course in self.__courses:
            print(f"Student {self.__name} has already enrolled in course {course}.")
        else:
            self.__courses.append(course)
```

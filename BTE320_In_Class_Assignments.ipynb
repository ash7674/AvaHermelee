```python
import random

play = input("Do you want to play (y/n): ")

while play == "y":
  p_move = input("Choose between rock, paper, scissors: ").lower()
  c_move = random.choice(["rock", "paper", "scissors"])

  print(f"Player move: {p_move}\nComputer move: {c_move}")

  if p_move == c_move:
    print("Draw")
  elif (
      (p_move == "rock" and c_move == "scissors")or
      (p_move == "scissors" and c_move == "paper") or
      (p_move == "paper" and c_move == "rock")
  ):
    print("Player wins!")
  else:
    print("Computer wins")

  play = input("Do you want to play again (y/n): ")
```

Append items to list iteratively (while loop). starting with an empty list, design a program in python that populates the list by repeatedly asking the user for numerical inputs. the process stops when the user gives 0. finally, the list should be displayed in descending order.

```python
numbers = []  # start with an empty list

while True:
    num = float(input("Enter a number (0 to stop): "))

    if num == 0:
        break
    numbers.append(num)

numbers.sort(reverse=True)

print("Numbers in descending order:", numbers)
```

product prices (dictionaries) a company sells two products: P-01 $9.95, P-02 $10.50. write a program that sets up a dictionary with product IDs as keys and prices as their values. Prompt the user to enter a product ID, then display that products price. sample run: Enter product ID: P-01
P-01 9.95. for your program with a dictionary of product IDs and their prices, refine it so that if the user enters a product ID not in the dictionary, it displays a message

```python
products = {
    "P-01": 9.95,
    "P-02": 10.50
}

product_id = input("Enter product ID: ")

if product_id in products:
    print(product_id, products[product_id])
else:
    print("Product ID not found.")
```

dynamic dictionary; N distinct key:value pairs (while loop/ if-elif-else statements). propulate an empty dictionary iteratively N times (choose your own value for N) with name:count pairs. Names (keys) are strings and counts (values) are integers. the keys must be given as inputs from the user on each iteration. use conditionals to check if the key has been added in the dictionary before. if not, set its count to 1; otherwise, increase its count by 1.

```python
names = {}

while True:
  name = input("Give a name: ")
  if name == "":
    break

  if name in names:
    names[name] += 1
  else:
    names[name] = 1

print(names)
```

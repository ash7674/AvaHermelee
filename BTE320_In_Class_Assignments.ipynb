```python
num1 = float(input("Enter first number: "))
op = input("Choose operator: ")
num2 = float(input("Enter second number: "))

if op == "+":
    res = num1 + num2

elif op == "-":
    res = num1 - num2

elif op == "*":
    res = num1 * num2

elif op == "/":
    res = num1 / num2

elif op == "**":
  res = num1 ** num2

elif op == "//":
    res = num1 // num2

elif op == "%":
  res = num1 % num2

else:
    print("Invalid operator")

print(f"Result: {res}")
```
